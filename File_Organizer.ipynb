```python
from logging import exception
import os ;
from urllib.parse import urlparse, unquote



def scanner_files(folderpath : str )->list :
    parsed = urlparse(folderpath)
    if parsed.scheme == "file":
        root_path = unquote(parsed.path)
    else:
        # Assume it's already a normal file system path
        root_path = folderpath

    files_info = []

    # --- 2. Walk through all folders and subfolders ---
    # os.walk goes through root_path and every subfolder inside it
    for current_path, dirs, files in os.walk(root_path):
        # --- 3. For each file, collect info into a dict ---
        for file_name in files:
            full_path = os.path.join(current_path, file_name)

            # File size in bytes
            size_in_bytes = os.path.getsize(full_path)

            # Split into base name and extension (".txt", ".py", etc.)
            _, ext = os.path.splitext(file_name)
            extension = ext.lstrip(".")  # remove the dot, e.g. ".txt" -> "txt"

            file_dict = {
                "name": file_name,
                "path": full_path,
                "size": size_in_bytes,
                "extension": extension
            }

            files_info.append(file_dict)

    return files_info

```
